# Cleaning the data

```
#Renaming the columns and making them similar

columns = ['PRODUCT', 'VALUE', 'DATE_TIME', 'CELL_ON_SITE', 'DW_NUMBER_INT', 'DW_B_NUMBER_INT', 'COUNTRY_A', 'COUNTRY_B', 'CELL_ID', 'SITE_ID']

CDR507.columns = columns
CDR507.head()
```

```
#Changing the columns of CDR508
CDR508.columns = columns

CDR508.head()
```

```
#Changing the columns of CDR509

CDR509.columns = columns

CDR509.head()
```

# Concatenating the three datasets

```
#Joining the three datasets
df_concat = pd.concat([CDR507, CDR508, CDR509], ignore_index = True)
df_concat.tail()
```

# Dropping the columns that are not useful

```
df_concat.drop(columns= ['COUNTRY_A', 'COUNTRY_B', 'CELL_ON_SITE'], inplace=True)
df_concat.head()
```

# Splitting the date_time column to date and time

```
#Splitting the date_time column and creating two new columns to hold the two variables 

df_concat[['date', 'time']] = df_concat.DATE_TIME.str.split(expand=True)
df_concat.head()
```

```
#Dropping DATE_TIME column
df_concat.drop(columns=['DATE_TIME'], inplace=True)
df_concat.head()
```

# Merging the concatenated dataframe with cells_geo

```
#Creating a variable to store the merged dataframe

mergd = cells_geo.merge(df_concat, how='left', on='SITE_ID')
mergd.head()
```

```
#Creating a variable to hold the values of the merged tabe where time is not null

mag = mergd[mergd.time.notnull()]
mag.head()
```
